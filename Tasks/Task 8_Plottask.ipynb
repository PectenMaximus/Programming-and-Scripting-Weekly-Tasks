### 1. A Histogram of a Normal Distribution 

```python
# Imports
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generate some data for this 
loc = 5.0 # Note to self Loc is the mean
scale= 2.0 # Note to self scale is the standard deviation 
data = np.random.normal(loc,scale, 1000)

# Fit a normal distribution to the data: mean and standard deviation
mu, std = norm.fit(data) 

# Plot the histogram
plt.hist(data, bins=25, fc="lightgreen", ec="darkgreen", density= True, alpha = 0.6)


# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

#Tidy up Plot
plt.xlabel("x")
plt.ylabel("y")
title = "Fit Values: {:.2f} and {:.2f}".format(mu, std) # Shows the acutal values for mean and standard deviation in the title 
plt.title(title)
plt.legend()
```

### 2. A plot of the function h(x)=x3

```python
# Define the range for x
x = np.linspace(0, 10, 100)

# Define the function h(x) = x^3
h_x = x**3

# Plot the function
plt.plot(x, h_x, label="h(x) = x*3", color="green")

# Add labels, title, and legend
plt.xlabel("x")
plt.ylabel("h(x)")
plt.title("Plot of h(x) = x*3")
plt.legend()

# Show the plot
plt.show()
```
