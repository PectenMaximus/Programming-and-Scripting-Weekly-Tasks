### 1. A Histogram of a Normal Distribution 

```python
# Imports
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Generate some data for this 
loc = 5.0 # Note to self Loc is the mean
scale= 2.0 # Note to self scale is the standard deviation 
data = np.random.normal(loc,scale, 1000)

# Fit a normal distribution tothe data:
# mean and standard deviation
mu, std = norm.fit(data) 

# Plot the histogram.
plt.hist(data, bins=25, fc="lightgreen", ec="darkgreen", density= True, alpha = 0.6)

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

plt.plot(x, p, 'k', linewidth=2)
title = "Fit Values: {:.2f} and {:.2f}".format(mu, std)
plt.title(title)
```

### 2. A plot of the function h(x)=x3

```python
# Define the function h(x) = x3
h_x = np.linspace(0, 10, 100)
h_y = h_x**3

# Plot the histogram and the function on the same axes
plt.hist(data, bins=25, fc="lightgreen", ec="darkgreen", density=True, alpha=0.6, label="Histogram")
plt.plot(x, p, 'k', linewidth=2, label="Normal Distribution Fit")
plt.plot(h_x, h_y, 'r--', linewidth=2, label="h(x) = x^3")

# Add labels, title, and legend
plt.xlabel("x")
plt.ylabel("Density / h(x)")
plt.title("Histogram, Normal Distribution Fit, and h(x) = x^3")
plt.legend()

# Show the plot
plt.show()
```
